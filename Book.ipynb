```python
import pandas as pd
```

```python
books_df= pd.read_csv("books.csv")
```

```python
books_df.head()
```

```python
books_df.info()
```

```python
books_df = books_df[['title','authors','categories','description']]
books_df.head()
```

```python
books_df['authors'] = books_df['authors'].apply(lambda x:str(x))
```

```python
books_df['authors'] = books_df['authors'].apply(lambda x:x.split())
```

```python
books_df
```

```python
books_df.dropna(inplace=True)
```

```python
books_df['description'] = books_df['description'].apply(lambda x:x.split())
books_df
```

```python
books_df['categories'] = books_df['categories'].apply(lambda x:x.split())
books_df
```

```python
books_df['tags'] =books_df['authors']+books_df['categories']+books_df['description'] 
```

```python
books_df
```

```python
new_df = books_df[['title','tags','categories']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 50000, stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
new_df
```

```python
def recommend(book,n):
    lst =[]
    book_index =new_df[new_df['title']==book].index[0]
    distances = similarity[book_index]
    book_list = sorted(list(enumerate(distances)),reverse= True,key = lambda x:x[1])[0:n+1]
    genre = new_df.iloc[book_list[0][0]].categories
    for i in book_list[1:]:
        a = new_df.iloc[i[0]].title
        b=0
        if genre[0] in new_df.iloc[i[0]].categories:
            b=1
        lst.append((a,b))
    result =[]
    for i in lst:
        if i[1]==1:
            result.append(i[0])
    for i in lst:
        if i[1]==0:
            result.append(i[0])
    print(result)
    
```

```python
recommend(input('Enter the book name to recommend:'),int(input('No of book to recommend :')))
```
